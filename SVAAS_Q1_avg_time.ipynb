```python
# Sorting the data
df1_new.sort_values(by=['created_date'], ascending=False, inplace=True)
```

```python
# Dropping duplicates
df2=df1_new.drop_duplicates(subset=['appointment_id','schedule_status'], keep='first')
```

```python
# how many status has completed
df2[df2['schedule_status']=='completed']
```

```python
df2[df2['appointment_id']==2849]
```

```python
# In main data
df1[df1['appointment_id']==2849]
```

```python
# Makining pivit
pivoted = df2.pivot(index="appointment_id", columns='schedule_status', values="created_date").reset_index()
pivoted
```

```python
# Adding columns complted, and consulation time
pivoted['complted_time']=pivoted['confirmed']-pivoted['booked']
pivoted['conslutation_time']=pivoted['completed']-pivoted['in progress']
```

```python
pivoted.head()
```

```python
# Makining pivit
pivoted1= df2.pivot(index=["appointment_id",'doctor_id'], columns='schedule_status', values="created_date").reset_index()
pivoted1.head()
```

```python
# Adding columns complted, and consulation time
pivoted1['complted_time']=pivoted1['confirmed']-pivoted1['booked']
pivoted1['conslutation_time']=pivoted1['completed']-pivoted1['in progress']
```

```python
pivoted1.head()
```

```python
#Doctor_id wise average completed time and avg consultaion time
df_final=pivoted1.groupby('doctor_id').agg({'complted_time':'mean','conslutation_time':'mean'})
```

```python
# Checking head of df_final
df_final.head()
```

```python
# saving file into csv 
df_final.to_csv('doctor_wise.csv',date_format='%Y-%m-%d %H:%M:%S')
```

```python
df_final1=df_final.copy()
```

```python
df_final1.info()
```

```python
# Extyracting the hours from timedelta64[ns]
df_final1['complted_time_hrs']=df_final1['complted_time'].dt.total_seconds()/3600
df_final1['conslutation_time_hrs']=df_final1['conslutation_time'].dt.total_seconds()/3600
```

```python
df_final1.head()
```

```python
# saving file into csv 
df_final1.to_csv('doctor_hrs_wise.csv',date_format='%H')
```
