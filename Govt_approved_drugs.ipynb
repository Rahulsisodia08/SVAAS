## Goverment approved drugs

```python
# Importing Libarary
import urllib.request
import pandas as pd
# To ignore warnings
import warnings
warnings.filterwarnings('ignore')
# For regex
import re
import numpy as np
# all rows in the dataframe
pd.set_option('display.max_rows', None)

# Url link
url = "https://cdscoonline.gov.in/CDSCO/soam/drug_data.jsp"

#Reading with HTML
with urllib.request.urlopen(url) as i:
    html = i.read()
    
data = pd.read_html(html)[0]



#Removing useless columns
data.drop(columns=[5,6,7,8,9,10,11,12],axis=1,inplace=True)
#Removing useless rows
data.drop([data.index[0], data.index[1]],inplace=True)
#droping index
data.reset_index(drop=True, inplace=True)
#Writting to csv
#data.to_csv('govt_approved_drug.csv', header=False, index=False)

# Making copy of data
df=data.copy()
#removing useless index
df.drop(df.index[0],inplace=True)
#renaming the columns
df.rename(columns={0:"Sr.No",1:"Drug Name",2:"Strength",3:"Indication",4:"Date of Approval"},inplace=True)
#resst the index
df.reset_index()
```

## Changing date format to MMM-YYYY

```python
# Changing to date time
df['Date of Approval']=pd.to_datetime(df['Date of Approval'].astype(str), errors='coerce').dt.date
# converting to base format '%Y-%m-%d'
df['Date of Approval'] = pd.to_datetime(df['Date of Approval'], format='%Y-%m-%d')
#converting to MMM-YYYY format
df['Date of Approval']=df['Date of Approval'].dt.strftime('%b-%y').str.capitalize()

# drug name lower case
df['Drug Name']=df['Drug Name'].str.lower()
```
