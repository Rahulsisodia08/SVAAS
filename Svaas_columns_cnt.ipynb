# SVAAS Columns Counts 
### 'complaints', 'diagnosis', 'advice', 'bp_dia', 'bp_sys', 'pulse', 'weight', 'height','bmi'

```python
import pandas as pd
import numpy as np
```

```python
# uPLOADING DATA
df= pd.read_excel('brdg_appt_conslt_mapping (1).xlsx')
df.head()
```

```python
df.columns
```

```python
df2=df.copy()
df2.head()
```

```python
df2=df2.fillna('Blank')
```

```python
# Columns counts on complaints
complaints_cnt=df2.groupby('complaints').agg({'complaints':'count'}).rename(columns={"complaints":"complaints_counts"})
 #df.C.isnull().groupby().sum().astype(int).reset_index(name='count')
# Head
complaints_cnt.head()
```

```python
# Saving files
complaints_cnt.to_excel('complaints_cnt.xlsx')
```

```python
# Columns counts on complaints
diagnosis_cnt=df2.groupby('diagnosis').agg({'diagnosis':'count'}).rename(columns={"diagnosis":"diagnosis_counts"})

advice_cnt=df2.groupby('advice').agg({'advice':'count'}).rename(columns={"advice":"advice_counts"})

bp_dia_cnt=df2.groupby('bp_dia').agg({'bp_dia':'count'}).rename(columns={"bp_dia":"bp_dia_counts"})

bp_sys_cnt=df2.groupby('bp_sys').agg({'bp_sys':'count'}).rename(columns={"bp_sys":"bp_sys_counts"})

pulse_cnt=df2.groupby('pulse').agg({'pulse':'count'}).rename(columns={"pulse":"pulse_counts"})

weight_cnt=df2.groupby('weight').agg({'weight':'count'}).rename(columns={"weight":"weight_counts"})

height_cnt=df2.groupby('height').agg({'height':'count'}).rename(columns={"height":"height_counts"})

bmi_cnt=df2.groupby('bmi').agg({'bmi':'count'}).rename(columns={"bmi":"bmi_counts"})
```

```python
# Saving files
diagnosis_cnt.to_excel('diagnosis_cnt.xlsx')

advice_cnt.to_excel('advice_cnt.xlsx')

bp_dia_cnt.to_excel('bp_dia_cnt.xlsx')

bp_sys_cnt.to_excel('bp_sys_cnt.xlsx')

pulse_cnt.to_excel('pulse_cnt.xlsx')

weight_cnt.to_excel('weight_cnt.xlsx')

height_cnt.to_excel('height_cnt.xlsx')

bmi_cnt.to_excel('bmi_cnt.xlsx')
```
